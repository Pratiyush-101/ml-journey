```python
import statistics

scores = [72, 45, 88, 91, 60, 55, 78, 83, 49, 95]

def summarize(scores):
  return{
      "average": sum(scores) / len(scores),
      "max": max(scores),
      "min": min(scores),
      "count": len(scores),
      "median": statistics.median(scores)
}
passing = [s for s in scores if s >= 60]
grade = ["A" if s >= 90 else "B" if s >=75 else "C" if s >=60 else "D" if s >=50 else "F" for s in scores]
grade_count = {g: grade.count(g) for g in set(grade)}
sorted_score = sorted(scores, reverse=True)
top3 = sorted_score[:3]
for rank, score in enumerate(top3, start=1):
    print(f"Rank {rank}: {score}")

print(summarize(scores))
print("Passing scores: ", passing)
print(sorted(list(zip(scores, grade))))
print(grade_count)
print("Top 3: ", sorted_score[:3])
print("Bottom 3: ", sorted_score[-3:])

# print(sorted(scores))
```

```python
import numpy as np

#fake house dataset
sizes = np.array([600, 800, 1000, 1200, 1500])
prices = np.array([300, 400, 500, 600, 750])

#model parameters
w = 0.5
b = 0

#predict
predicted = w * sizes + b

# coumpute loss
mse = np.mean((predicted - prices) ** 2)

print(f"Predicted: {predicted}")
print(f"Actual: {prices}")
print(f"MSE Loss: {mse:.2f}")
```

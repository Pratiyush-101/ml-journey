```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print(a+b)
print(a*b)
print(a ** 2)
print(np.dot(a, b))
print(np.mean(a))
print(np.std(a))

normalized = (a - np.mean(a)) / np.std(a) #Important
print(normalized)
print(np.mean(normalized))
print(np.std(normalized))
```
